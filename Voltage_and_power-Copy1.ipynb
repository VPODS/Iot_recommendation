```python
import json
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
f = open('Voltage_Power.json')
 
# returns JSON object as
# a dictionary
data = json.loads(f.read())
```

```python
df = pd.json_normalize(data)
```

```python
energy_mean = df['telemetry.EnergyAmountkWh'].mean()
energy_mean
```

```python
Energy_desc = []

for i in df['telemetry.EnergyAmountkWh'].values:
    if i > energy_mean+10:
        Energy_desc.append("Energy is high")
    elif i < energy_mean-10:
        Energy_desc.append("Energy is low")
    else:
        Energy_desc.append("Energy is normal")
```

```python
df['Energy_desc'] = Energy_desc
```

```python
df
```

```python
voltage_mean = df['telemetry.VoltageReading'].mean()
voltage_mean
```

```python
voltage_desc = []

for i in df['telemetry.VoltageReading'].values:
    if i > voltage_mean+10:
        voltage_desc.append("voltage is high")
    elif i < voltage_mean-10:
        voltage_desc.append("voltage is low")
    else:
        voltage_desc.append("voltage is normal")
```

```python
df['voltage_desc'] = voltage_desc
```

```python
power_mean = df['telemetry.PowerReading'].mean()
power_mean
```

```python
power_desc = []

for i in df['telemetry.PowerReading'].values:
    if i > power_mean+10:
        power_desc.append("power is high")
    elif i < power_mean-10:
        power_desc.append("power is low")
    else:
        power_desc.append("power is normal")
```

```python
df['power_desc']= power_desc
```
