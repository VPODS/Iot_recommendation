```python
import json
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
f = open('Temperature_pressure.json')
 
# returns JSON object as
# a dictionary
data = json.loads(f.read())
```

```python
data
```

```python
df = pd.json_normalize(data)
```

```python
df
```

```python
df.dtypes
```

```python
for i in df.columns:
    print(i)
    print(df[i].unique())
```

```python
for i in df.columns:
    print(i)
    print(df[i].value_counts())
```

```python
corr = df.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(df.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.columns)
plt.show()
```

```python

#plot the scatter plot of balance and salary variable in data
plt.scatter(df['messageProperties.iothub-creation-time-utc'],df['telemetry.Temperature'])
plt.show()

```

```python
#plot the scatter plot of balance and age variable in data
df.plot.scatter(x="messageProperties.iothub-creation-time-utc",y="telemetry.Pressure")
plt.show()
```

```python
import seaborn as sns
#plot the pair plot of salary, balance and age in data dataframe.
sns.pairplot(data = df, vars=['telemetry.Relative_Humidity','telemetry.Pressure','telemetry.Temperature'])
plt.show()
```

```python
# Creating a matrix using age, salry, balance as rows and columns
df[['telemetry.Relative_Humidity','telemetry.Pressure','telemetry.Temperature']].corr()

#plot the correlation matrix of salary, balance and age in data dataframe.
sns.heatmap(df[['telemetry.Relative_Humidity','telemetry.Pressure','telemetry.Temperature']].corr(), annot=True, cmap = 'Reds')
plt.show()
```
